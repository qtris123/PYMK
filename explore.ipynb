```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("pypiahmad/social-circles")

print("Path to dataset files:", path)
```

```python
import os
import shutil
# 2. Choose your destination folder
dst = "C:\qtri\Self-learning\ML System Sidequest\People You May Know\data"

# 3. Create folder if needed
os.makedirs(dst, exist_ok=True)


# 3. Copy everything
for name in os.listdir(path):
    src_item = os.path.join(path, name)
    dst_item = os.path.join(dst, name)

    # If directory → copy tree (merge allowed)
    if os.path.isdir(src_item):
        # if directory already exists, remove then copy
        if os.path.exists(dst_item):
            shutil.rmtree(dst_item)
        shutil.copytree(src_item, dst_item)
    else:
        # If file → copy normally
        shutil.copy2(src_item, dst_item)

print("Dataset fully copied to:", dst)
```
